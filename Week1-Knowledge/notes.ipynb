## 1. Introduction: Reasoning from Knowledge

Humans use **knowledge** to reason and draw new conclusions.
AI systems can do the same ‚Äî these are called **Knowledge-Based Agents**.

> üß© Example:
>
> * If it didn‚Äôt rain, Harry visited Hagrid today.
> * Harry visited Hagrid or Dumbledore today, but not both.
> * Harry visited Dumbledore today.
>
> From these, we can reason:
> ‚Üí Harry did **not** visit Hagrid.
> ‚Üí Therefore, **it rained today**.

AI represents and manipulates such knowledge using **logic**.

---

## 2. üßæ Sentences and Propositional Logic

### ‚û§ Sentence

An **assertion** about the world, expressed in a knowledge representation language.

### ‚û§ Propositional Logic

A way of representing facts about the world using **propositions**, which can be **True (T)** or **False (F)**.

---

## 3. üî† Propositional Symbols

Usually capital letters (`P`, `Q`, `R`) represent simple propositions.

Example:

* P: ‚ÄúIt is raining.‚Äù
* Q: ‚ÄúI am indoors.‚Äù

---

## 4. üîó Logical Connectives

| Symbol | Name                   | Meaning              | True When                  |
| :----- | :--------------------- | :------------------- | :------------------------- |
| ¬¨      | **Not**                | Negates truth value  | Opposite of P              |
| ‚àß      | **And**                | Both must be true    | P and Q both true          |
| ‚à®      | **Or**                 | At least one true    | P or Q or both true        |
| ‚äï      | **Exclusive Or (XOR)** | Exactly one true     | Only P or only Q           |
| ‚Üí      | **Implication**        | ‚ÄúIf P, then Q‚Äù       | False only when P=T, Q=F   |
| ‚Üî      | **Biconditional**      | ‚ÄúP if and only if Q‚Äù | True when P and Q are same |

---

## 5. ‚ùó Truth Tables

### (a) **NOT (¬¨)**

|  P  |  ¬¨P |
| :-: | :-: |
|  T  |  F  |
|  F  |  T  |

---

### (b) **AND (‚àß)**

|  P  |  Q  | P ‚àß Q |
| :-: | :-: | :---: |
|  F  |  F  |   F   |
|  F  |  T  |   F   |
|  T  |  F  |   F   |
|  T  |  T  |   T   |

---

### (c) **OR (‚à®)**

|  P  |  Q  | P ‚à® Q |
| :-: | :-: | :---: |
|  F  |  F  |   F   |
|  F  |  T  |   T   |
|  T  |  F  |   T   |
|  T  |  T  |   T   |

üëâ *Inclusive* OR allows both to be true.
üëâ *Exclusive* OR (XOR) is true only if **exactly one** is true.

---

## 6. ‚û°Ô∏è **Implication (‚Üí)**

### P ‚Üí Q means:

> ‚ÄúIf P, then Q.‚Äù

üìä **Truth Table for Implication**

|  P  |  Q  | P ‚Üí Q | Explanation                                                                                         |
| :-: | :-: | :---: | :-------------------------------------------------------------------------------------------------- |
|  T  |  T  | **T** | If P is true and Q is true, the rule ‚Äúif P then Q‚Äù is satisfied.                                    |
|  T  |  F  | **F** | If P is true but Q is false, the rule is broken ‚Äî this is the only case where implication is false. |
|  F  |  T  | **T** | If P is false, we don‚Äôt care what Q is ‚Äî the rule isn‚Äôt violated.                                   |
|  F  |  F  | **T** | Same here ‚Äî if P is false, ‚Äúif P then Q‚Äù doesn‚Äôt get tested, so it‚Äôs considered true.               |

### üí° Example

If P = ‚ÄúIt‚Äôs raining‚Äù and Q = ‚ÄúI‚Äôm indoors‚Äù,
then **P ‚Üí Q** means ‚ÄúIf it‚Äôs raining, then I‚Äôm indoors.‚Äù

---

## 7. ‚Üî **Biconditional (If and Only If)**

### P ‚Üî Q

or
‚Äúif and only if‚Äù
OR
**P if and only if Q**

üß† **What It Means**
A biconditional says:

> ‚ÄúP and Q have the same truth value.‚Äù

üìä **Truth Table for Biconditional**

|  P  |  Q  | P ‚Üî Q | Explanation              |
| :-: | :-: | :---: | :----------------------- |
|  T  |  T  | **T** | Both are true ‚Üí match ‚úÖ  |
|  T  |  F  | **F** | Don‚Äôt match ‚ùå            |
|  F  |  T  | **F** | Don‚Äôt match ‚ùå            |
|  F  |  F  | **T** | Both are false ‚Üí match ‚úÖ |

### üí° Example

If P = ‚ÄúIt‚Äôs raining‚Äù and Q = ‚ÄúI‚Äôm indoors,‚Äù
then **P ‚Üî Q** means:

> ‚ÄúIf it‚Äôs raining, then I‚Äôm indoors, and if I‚Äôm indoors, then it‚Äôs raining.‚Äù

---

## 8. üß© Model

A **model** is a specific assignment of truth values to all propositions.

Example:

* P: ‚ÄúIt‚Äôs raining.‚Äù
* Q: ‚ÄúIt‚Äôs Tuesday.‚Äù

One possible model:
‚Üí {P = True, Q = False}

With 2 propositions, there are 2¬≤ = 4 possible models.

---

## 9. üß† Knowledge Base (KB)

A **knowledge base** is a collection of sentences (facts, rules) known to the AI.

Example:

```
1. If it didn‚Äôt rain, Harry visited Hagrid.
2. Harry visited Hagrid or Dumbledore, but not both.
3. Harry visited Dumbledore.
```

From these, KB can infer:
‚Üí It rained today.

---

## 10. ‚ä® **Entailment**

**Entailment (‚ä®)** means:

> If Œ± ‚ä® Œ≤, then in *every model* where Œ± is true, Œ≤ is also true.

### Example:

* Œ±: ‚ÄúIt‚Äôs Tuesday in January.‚Äù
* Œ≤: ‚ÄúIt‚Äôs January.‚Äù
  Then Œ± ‚ä® Œ≤ (every world where Œ± is true, Œ≤ is also true).

üß© Difference:

* **Implication (‚Üí)** is a **logical connective** inside sentences.
* **Entailment (‚ä®)** is a **relationship** between two sentences.

# Propositional Logic: Inference & Knowledge Engineering

## üìñ Part 1: Inference

### What is Inference?
**Inference** is the process of deriving new sentences (conclusions) from old ones (premises).

**Example**: In Harry Potter logic:
- From sentences 1, 2, and 3 ‚Üí we can infer sentences 4 and 5

---

## üîç Model Checking Algorithm

### Definition
To determine if **KB ‚ä® Œ±** (Does the knowledge base entail Œ±?):

**Algorithm Steps:**
1. **Enumerate all possible models** (all truth value combinations)
2. **Check each model**: If KB is true in that model, is Œ± also true?
3. **Conclusion**: If Œ± is true in EVERY model where KB is true ‚Üí KB ‚ä® Œ±

---

### üìä Example Problem

**Given:**
- **P**: It is a Tuesday
- **Q**: It is raining
- **R**: Harry will go for a run

**Knowledge Base (KB):**
1. `(P ‚àß ¬¨Q) ‚Üí R` ‚Äî "If it's Tuesday AND not raining, Harry goes for a run"
2. `P` ‚Äî "It IS Tuesday"
3. `¬¨Q` ‚Äî "It is NOT raining"

**Query**: `R` ‚Äî Does Harry go for a run?

### Step-by-Step Evaluation

#### Step 1: Apply `P` is true
Since P must be true, KB is **false** in all models where P = false

#### Step 2: Apply `¬¨Q` (Q is false)
Since Q must be false, KB is **false** in all models where Q = true

#### Step 3: Apply `(P ‚àß ¬¨Q) ‚Üí R`
When P = true AND Q = false, R must be true
- Model where R = false ‚Üí KB is **false**
- Model where R = true ‚Üí KB is **true** ‚úì

## üíª Code Representation

### Example: Harry Potter Logic Puzzle

```python
from logic import *

# Create symbols for each proposition
rain = Symbol("rain")           # It is raining
hagrid = Symbol("hagrid")       # Harry visited Hagrid
dumbledore = Symbol("dumbledore")  # Harry visited Dumbledore

# Build the Knowledge Base
knowledge = And(
    Implication(Not(rain), hagrid),      # ¬¨rain ‚Üí hagrid
    Or(hagrid, dumbledore),              # hagrid ‚à® dumbledore
    Not(And(hagrid, dumbledore)),        # ¬¨(hagrid ‚àß dumbledore)
    dumbledore                           # dumbledore is TRUE
)
```

**What the KB says:**
1. If it's not raining, Harry visited Hagrid
2. Harry visited Hagrid OR Dumbledore (at least one)
3. Harry did NOT visit both Hagrid and Dumbledore
4. Harry DID visit Dumbledore (fact)

## üîÑ The check_all Function

### Required Inputs
1. **Knowledge Base** ‚Äî Used to draw inferences
2. **Query** ‚Äî The proposition we want to check
3. **Symbols** ‚Äî List of all atomic propositions (variables)
4. **Model** ‚Äî Assignment of truth values to symbols

---

### Algorithm Code

```python
def check_all(knowledge, query, symbols, model):
    
    # Base case: all symbols have been assigned
    if not symbols:
        # If KB is true in this model
        if knowledge.evaluate(model):
            return query.evaluate(model)  # Check if query is true
        return True  # KB is false ‚Üí vacuous truth
    
    # Recursive case: still have symbols to assign
    else:
        remaining = symbols.copy()
        p = remaining.pop()  # Pick one symbol
        
        # Create two models
        model_true = model.copy()
        model_true[p] = True
        
        model_false = model.copy()
        model_false[p] = False
        
        # Recurse on both and combine with AND
        return (check_all(knowledge, query, remaining, model_true) and
                check_all(knowledge, query, remaining, model_false))
```

---

### üéØ Key Insight: Why "Vacuous Truth"?

**We only care about models where KB is true!**

**Example**: 
- **KB**: P (Harry plays seeker), Q (Oliver plays keeper), (P ‚àß Q) ‚Üí R (Gryffindor wins)
- **Model**: ¬¨P (Harry plays beater instead)

In this model, we don't care about R because our KB says P is true. This model is **irrelevant** to our knowledge.

**If KB is false in a model ‚Üí return True (ignore this model)**

## üéÆ Example 1: Clue/Cluedo Game

### Game Setup
**Murder mystery with:**
- **People**: Mustard, Plum, Scarlet
- **Tools**: Knife, Revolver, Wrench
- **Locations**: Ballroom, Kitchen, Library

**Goal**: Determine WHO did it, with WHAT tool, in WHICH location

---

### Building the Knowledge Base

#### Step 1: Game Rules
One person, one tool, one location must be true:

```python
knowledge = And(
    # Exactly one person is guilty
    Or(mustard, plum, scarlet),
    
    # Exactly one tool was used
    Or(knife, revolver, wrench),
    
    # Exactly one location
    Or(ballroom, kitchen, library)
)
```

---

#### Step 2: Initial Cards (What We See)
We see: Mustard, Kitchen, Revolver
Therefore, these are NOT the answer:

```python
knowledge = And(
    # ... previous rules ...
    Not(mustard),
    Not(kitchen),
    Not(revolver)
)
```

---

#### Step 3: False Guess
Someone guesses: "Scarlet, Library, Wrench" ‚Äî **WRONG**
This means: At least ONE of these is false

```python
knowledge = And(
    # ... previous knowledge ...
    Or(Not(scarlet), Not(library), Not(wrench))
)
```

**Translation**: ¬¨Scarlet ‚à® ¬¨Library ‚à® ¬¨Wrench
(At least one must be false)

---

#### Step 4: More Clues
Someone shows us Plum card:

```python
knowledge = And(
    # ... previous knowledge ...
    Not(plum)
)
```

**Now we can deduce**: Since Mustard and Plum are out ‚Üí **Scarlet is guilty!**

---

#### Step 5: Final Clue
We learn: NOT the Ballroom

```python
knowledge = And(
    # ... previous knowledge ...
    Not(ballroom)
)
```

---

### üß© Final Deduction

**What we know:**
1. Must be Ballroom, Kitchen, or Library
2. NOT Ballroom, NOT Kitchen
3. Therefore: **Library ‚úì**

**From the false guess:**
- Scarlet ‚úì, Library ‚úì, Wrench ?
- Since at least one must be false, and we know Scarlet and Library are TRUE
- Therefore: **Wrench is FALSE**, so it must be **Knife ‚úì**

**Solution: Scarlet, Knife, Library** üéâ

---

### Complete Clue Code (`clue.py`)

```python
import termcolor
from logic import *

# Define symbols for characters, rooms, and weapons
mustard = Symbol("ColMustard")
plum = Symbol("ProfPlum")
scarlet = Symbol("MsScarlet")
characters = [mustard, plum, scarlet]

ballroom = Symbol("ballroom")
kitchen = Symbol("kitchen")
library = Symbol("library")
rooms = [ballroom, kitchen, library]

knife = Symbol("knife")
revolver = Symbol("revolver")
wrench = Symbol("wrench")
weapons = [knife, revolver, wrench]

symbols = characters + rooms + weapons

def check_knowledge(knowledge):
    """Check what we can deduce about each symbol"""
    for symbol in symbols:
        if model_check(knowledge, symbol):
            # We can prove this symbol is TRUE
            termcolor.cprint(f"{symbol}: YES", "green")
        elif not model_check(knowledge, Not(symbol)):
            # We can't prove it's false, so it's uncertain
            print(f"{symbol}: MAYBE")
        # If neither condition is true, it's proven FALSE (print nothing)

# Build the knowledge base
knowledge = And(
    # Game rules: one person, one room, one weapon
    Or(mustard, plum, scarlet),
    Or(ballroom, kitchen, library),
    Or(knife, revolver, wrench)
)

# Initial cards we saw (these are NOT the answer)
knowledge.add(And(
    Not(mustard), Not(kitchen), Not(revolver)
))

# Someone's wrong guess (at least one is false)
knowledge.add(Or(
    Not(scarlet), Not(library), Not(wrench)
))

# Additional cards revealed
knowledge.add(Not(plum))
knowledge.add(Not(ballroom))

# Check what we can deduce
check_knowledge(knowledge)
```

**Output:**
```
MsScarlet: YES     (proven guilty!)
library: YES       (proven location!)
knife: YES         (proven weapon!)
```

**Key Feature**: The `check_knowledge()` function has three outcomes:
- **YES** (green): We can prove it's true ‚Üí `model_check(KB, symbol)` returns True
- **MAYBE**: We can't prove it's false ‚Üí `model_check(KB, Not(symbol))` returns False
- **Nothing printed**: We can prove it's false ‚Üí Both checks return False

## üéì Example 2: House Assignment Puzzle (`puzzle.py`)

### Problem
Four people (Gilderoy, Pomona, Minerva, Horace) assigned to four houses (Gryffindor, Hufflepuff, Ravenclaw, Slytherin). Each person in exactly one house.

---

### Complete Implementation

```python
from logic import *

people = ["Gilderoy", "Pomona", "Minerva", "Horace"]
houses = ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]

symbols = []
knowledge = And()

# Create all 16 propositions (4 people √ó 4 houses)
for person in people:
    for house in houses:
        symbols.append(Symbol(f"{person}{house}"))

# CONSTRAINT 1: Each person belongs to a house
for person in people:
    knowledge.add(Or(
        Symbol(f"{person}Gryffindor"),
        Symbol(f"{person}Hufflepuff"),
        Symbol(f"{person}Ravenclaw"),
        Symbol(f"{person}Slytherin")
    ))

# CONSTRAINT 2: Only one house per person
for person in people:
    for h1 in houses:
        for h2 in houses:
            if h1 != h2:
                # If person is in h1, they're NOT in h2
                knowledge.add(
                    Implication(Symbol(f"{person}{h1}"), 
                              Not(Symbol(f"{person}{h2}")))
                )

# CONSTRAINT 3: Only one person per house
for house in houses:
    for p1 in people:
        for p2 in people:
            if p1 != p2:
                # If p1 is in this house, p2 is NOT
                knowledge.add(
                    Implication(Symbol(f"{p1}{house}"), 
                              Not(Symbol(f"{p2}{house}")))
                )

# CLUE 1: Gilderoy is in Gryffindor or Ravenclaw
knowledge.add(
    Or(Symbol("GilderoyGryffindor"), Symbol("GilderoyRavenclaw"))
)

# CLUE 2: Pomona is NOT in Slytherin
knowledge.add(
    Not(Symbol("PomonaSlytherin"))
)

# CLUE 3: Minerva IS in Gryffindor
knowledge.add(
    Symbol("MinervaGryffindor")
)

# Find all true assignments
for symbol in symbols:
    if model_check(knowledge, symbol):
        print(symbol)
```

**Output Example:**
```
MinervaGryffindor    (given as fact)
GilderoyRavenclaw    (deduced from clues)
PomonaHufflepuff     (deduced by elimination)
HoraceSlytherin      (deduced by elimination)
```

---

### Challenge: Representing in Propositional Logic

**Need 16 propositions:**
- MinervaGryffindor, MinervaHufflepuff, MinervaRavenclaw, MinervaSlytherin
- PomonaGryffindor, PomonaHufflepuff, ...
- (4 people √ó 4 houses = 16 propositions)

---

### Constraints to Encode

#### 1. Each person belongs to A house:
```python
# For each person, at least one house must be true
Or(MinervaGryffindor, MinervaHufflepuff, 
   MinervaRavenclaw, MinervaSlytherin)
```
Repeat for each person (4 times)

#### 2. Each person belongs to ONLY ONE house:
```python
# For each pair of houses, if person in h1, NOT in h2
for h1 in houses:
    for h2 in houses:
        if h1 != h2:
            Implication(Symbol(f"{person}{h1}"), 
                       Not(Symbol(f"{person}{h2}")))
```
For each person: 12 implications (4 √ó 3)
Total: 48 implications for all 4 people

#### 3. Each house has ONLY ONE person:
```python
# For each pair of people, if p1 in house, p2 NOT in house
for p1 in people:
    for p2 in people:
        if p1 != p2:
            Implication(Symbol(f"{p1}{house}"), 
                       Not(Symbol(f"{p2}{house}")))
```
For each house: 12 implications (4 √ó 3)
Total: 48 more implications for all 4 houses

**Total Constraints**: 4 (at least one house) + 48 (one house per person) + 48 (one person per house) = **100 logical statements**

**Problem**: Very cumbersome! This is why **First-Order Logic** is useful (covered later).

## üé® Example 3: Mastermind Game (`mastermind.py`)

### Game Rules
- Player 1: Arranges colors in secret order
- Player 2: Guesses the order
- Player 1: Returns number of correct positions

---

### Complete Implementation

```python
from logic import *

colors = ["red", "blue", "green", "yellow"]
symbols = []

# Create all propositions: color + position
for i in range(4):
    for color in colors:
        symbols.append(Symbol(f"{color}{i}"))

knowledge = And()

# CONSTRAINT 1: Each color has a position
for color in colors:
    knowledge.add(Or(
        Symbol(f"{color}0"),
        Symbol(f"{color}1"),
        Symbol(f"{color}2"),
        Symbol(f"{color}3")
    ))

# CONSTRAINT 2: Only one position per color
for color in colors:
    for i in range(4):
        for j in range(4):
            if i != j:
                knowledge.add(Implication(
                    Symbol(f"{color}{i}"), 
                    Not(Symbol(f"{color}{j}"))
                ))

# CONSTRAINT 3: Only one color per position
for i in range(4):
    for c1 in colors:
        for c2 in colors:
            if c1 != c2:
                knowledge.add(Implication(
                    Symbol(f"{c1}{i}"), 
                    Not(Symbol(f"{c2}{i}"))
                ))

# CLUE 1: First guess - exactly 2 positions correct
# red0, blue1, green2, yellow3 ‚Üí 2 correct
# This means exactly 2 of these are true
knowledge.add(Or(
    And(Symbol("red0"), Symbol("blue1"), 
        Not(Symbol("green2")), Not(Symbol("yellow3"))),
    And(Symbol("red0"), Symbol("green2"), 
        Not(Symbol("blue1")), Not(Symbol("yellow3"))),
    And(Symbol("red0"), Symbol("yellow3"), 
        Not(Symbol("blue1")), Not(Symbol("green2"))),
    And(Symbol("blue1"), Symbol("green2"), 
        Not(Symbol("red0")), Not(Symbol("yellow3"))),
    And(Symbol("blue1"), Symbol("yellow3"), 
        Not(Symbol("red0")), Not(Symbol("green2"))),
    And(Symbol("green2"), Symbol("yellow3"), 
        Not(Symbol("red0")), Not(Symbol("blue1")))
))

# CLUE 2: Second guess - 0 correct
# blue0, red1, green2, yellow3 ‚Üí all wrong
knowledge.add(And(
    Not(Symbol("blue0")),
    Not(Symbol("red1")),
    Not(Symbol("green2")),
    Not(Symbol("yellow3"))
))

# Find the solution
for symbol in symbols:
    if model_check(knowledge, symbol):
        print(symbol)
```

**Output:**

# Inference Rules

## ‚ö° Why Inference Rules?

### The Problem with Model Checking
**Model Checking is INEFFICIENT** because it must:
- Consider EVERY possible model (2‚Åø combinations)
- Check if KB is true in each model
- Verify if query is true in all models where KB is true

**Example**: 20 variables ‚Üí 1,048,576 models to check!

---

### The Solution: Inference Rules
**Inference rules** allow us to generate new information based on existing knowledge **WITHOUT** considering every possible model.

**Structure of an Inference Rule:**
```
         Premise (what we know)
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
         Conclusion (what we can deduce)
```

- **Top (Premise)**: Our existing knowledge
- **Bottom (Conclusion)**: New knowledge we can generate

---

## üìã The Inference Rules

### 1Ô∏è‚É£ Modus Ponens

**Rule**: If we know an implication and its antecedent are true, then the consequent must be true.

**Formula:**
```
    Œ± ‚Üí Œ≤,  Œ±
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
         Œ≤
```

**Example:**
- **Premise 1**: If it is raining ‚Üí Harry is inside
- **Premise 2**: It IS raining
- **Conclusion**: Harry is inside ‚úì

**In symbols:**
- Know: `Rain ‚Üí Inside`
- Know: `Rain`
- Conclude: `Inside`

üì∏ **[IMAGE: Modus Ponens diagram]**

---

### 2Ô∏è‚É£ And Elimination

**Rule**: If an And proposition is true, then any individual component within it is true.

**Formula:**
```
      Œ± ‚àß Œ≤
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        Œ±
```
(Or we can conclude Œ≤)

**Example:**
- **Premise**: Harry is friends with Ron AND Hermione
- **Conclusion**: Harry is friends with Hermione ‚úì

**In symbols:**
- Know: `FriendsWithRon ‚àß FriendsWithHermione`
- Conclude: `FriendsWithHermione` (or `FriendsWithRon`)

üì∏ **[IMAGE: And Elimination diagram]**

**Why this works**: If both things are true together, each individual thing must be true!

---

### 3Ô∏è‚É£ Double Negation Elimination

**Rule**: A proposition that is negated twice is simply true (negatives cancel out).

**Formula:**
```
      ¬¨(¬¨Œ±)
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        Œ±
```

**Example:**
- **Premise**: "It is NOT true that Harry did NOT pass the test"
- Parsing: ¬¨(Harry did not pass) = ¬¨(¬¨(Harry passed))
- **Conclusion**: Harry passed the test ‚úì

**In symbols:**
- Know: `¬¨(¬¨Passed)`
- Conclude: `Passed`

üì∏ **[IMAGE: Double Negation Elimination diagram]**

**Think of it like math**: Two negatives make a positive!

---

### 4Ô∏è‚É£ Implication Elimination

**Rule**: An implication can be converted to an Or statement with the negated antecedent.

**Formula:**
```
      Œ± ‚Üí Œ≤
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    ¬¨Œ± ‚à® Œ≤
```

**Example:**
- **Premise**: "If it is raining, Harry is inside"
- **Conclusion**: "(It is NOT raining) OR (Harry is inside)" ‚úì

**In symbols:**
- Know: `Rain ‚Üí Inside`
- Conclude: `¬¨Rain ‚à® Inside`

üì∏ **[IMAGE: Implication Elimination diagram]**

---

#### Why Does This Work? ü§î

**Truth Table Proof:**

| P | Q | P ‚Üí Q | ¬¨P ‚à® Q |
|---|---|-------|--------|
| F | F | **T** | **T** |
| F | T | **T** | **T** |
| T | F | **F** | **F** |
| T | T | **T** | **T** |

Since `P ‚Üí Q` and `¬¨P ‚à® Q` have **identical truth values** in all cases, they are **logically equivalent**!

**Intuitive Explanation:**

An implication is true in two cases:

1. **Antecedent is false** ‚Üí Implication is trivially true (vacuous truth)
   - Represented by `¬¨P` in `¬¨P ‚à® Q`
   - If P is false, the whole Or is true

2. **Both are true** ‚Üí Implication is true
   - If P is true and Q is true ‚Üí `¬¨P ‚à® Q` is true (because Q is true)
   
3. **P true, Q false** ‚Üí Implication is false
   - `¬¨P ‚à® Q` is also false (¬¨P is false, Q is false)

---

### 5Ô∏è‚É£ Biconditional Elimination

**Rule**: A biconditional (if and only if) can be split into two implications joined by And.

**Formula:**
```
       Œ± ‚áî Œ≤
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    (Œ± ‚Üí Œ≤) ‚àß (Œ≤ ‚Üí Œ±)
```

**Example:**
- **Premise**: "It is raining IF AND ONLY IF Harry is inside"
- **Conclusion**: ("If raining ‚Üí Harry inside" AND "If Harry inside ‚Üí raining") ‚úì

**In symbols:**
- Know: `Rain ‚áî Inside`
- Conclude: `(Rain ‚Üí Inside) ‚àß (Inside ‚Üí Rain)`

üì∏ **[IMAGE: Biconditional Elimination diagram]**

**Why this works**: "If and only if" means the implication works BOTH ways!

---

### 6Ô∏è‚É£ De Morgan's Law

**Rule 1**: The negation of an And can be converted to an Or of negations.

**Formula:**
```
     ¬¨(Œ± ‚àß Œ≤)
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    ¬¨Œ± ‚à® ¬¨Œ≤
```

**Example:**
- **Premise**: "It is NOT true that BOTH Harry AND Ron passed the test"
- **Conclusion**: "(Harry did NOT pass) OR (Ron did NOT pass)" ‚úì

**In symbols:**
- Know: `¬¨(HarryPassed ‚àß RonPassed)`
- Conclude: `¬¨HarryPassed ‚à® ¬¨RonPassed`

üì∏ **[IMAGE: De Morgan's Law 1 diagram]**

**Intuition**: If they didn't BOTH pass, then at least one of them didn't pass!

---

**Rule 2**: The negation of an Or can be converted to an And of negations.

**Formula:**
```
     ¬¨(Œ± ‚à® Œ≤)
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    ¬¨Œ± ‚àß ¬¨Œ≤
```

**Example:**
- **Premise**: "It is NOT true that Harry OR Ron passed the test"
- **Conclusion**: "(Harry did NOT pass) AND (Ron did NOT pass)" ‚úì

**In symbols:**
- Know: `¬¨(HarryPassed ‚à® RonPassed)`
- Conclude: `¬¨HarryPassed ‚àß ¬¨RonPassed`

üì∏ **[IMAGE: De Morgan's Law 2 diagram]**

**Intuition**: If neither passed, then both of them didn't pass!

---

### De Morgan's Laws Summary

**Two forms:**
1. `¬¨(Œ± ‚àß Œ≤) ‚â° (¬¨Œ± ‚à® ¬¨Œ≤)` ‚Äî "Not both" = "At least one didn't"
2. `¬¨(Œ± ‚à® Œ≤) ‚â° (¬¨Œ± ‚àß ¬¨Œ≤)` ‚Äî "Not either" = "Both didn't"

**Memory trick**: The negation flips the operator (And ‚Üî Or) and distributes to both sides!

---

### 7Ô∏è‚É£ Distributive Property

**Rule 1**: And can distribute over Or.

**Formula:**
```
    Œ± ‚àß (Œ≤ ‚à® Œ≥)
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    (Œ± ‚àß Œ≤) ‚à® (Œ± ‚àß Œ≥)
```

**Example:**
- **Premise**: "Harry is brave AND (Ron is funny OR Hermione is smart)"
- **Conclusion**: "(Harry is brave AND Ron is funny) OR (Harry is brave AND Hermione is smart)" ‚úì

üì∏ **[IMAGE: Distributive Property 1 diagram]**

**Think algebra**: Like `a √ó (b + c) = (a √ó b) + (a √ó c)`

---

**Rule 2**: Or can distribute over And.

**Formula:**
```
    Œ± ‚à® (Œ≤ ‚àß Œ≥)
    ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    (Œ± ‚à® Œ≤) ‚àß (Œ± ‚à® Œ≥)
```

**Example:**
- **Premise**: "Harry passed OR (Ron failed AND Hermione failed)"
- **Conclusion**: "(Harry passed OR Ron failed) AND (Harry passed OR Hermione failed)" ‚úì

üì∏ **[IMAGE: Distributive Property 2 diagram]**

# Resolution & First-Order Logic

## üéØ Resolution: A Powerful Inference Rule

### What is Resolution?

**Resolution** is an inference rule that states: If one of two atomic propositions in a disjunction (Or) is false, the other must be true.

**Basic Formula:**
$$\frac{P \lor Q, \quad \neg P}{Q}$$

**Simple Rule**: "One or the other" + "Not the first" = "Must be the second"

---

### Basic Example

**Given:**
- "Ron is in the Great Hall OR Hermione is in the library"
- "Ron is NOT in the Great Hall"

**Conclude:**
- "Hermione is in the library" ‚úì

**Symbolic Notation:**
$$\frac{\text{RonHall} \lor \text{HermioneLibrary}, \quad \neg\text{RonHall}}{\text{HermioneLibrary}}$$

üì∏ **[IMAGE: Resolution rule showing simple case with two clauses]**

**Why it works**: If it must be one or the other, and we know it's not the first, then it MUST be the second!

---

## üîó Complementary Literals

**Definition**: Two of the same atomic propositions where one is negated and the other is not.

**Examples:**
- $P$ and $\neg P$
- $\text{Raining}$ and $\neg\text{Raining}$
- $\text{AtHome}$ and $\neg\text{AtHome}$

**Key Property**: Complementary literals are the "key" that unlocks resolution!

---

## üöÄ Generalized Resolution

### Full Resolution Rule

**Formula:**
$$\frac{P \lor Q, \quad \neg P \lor R}{Q \lor R}$$

**Meaning**: If we have "P or Q" and "not P or R", we can conclude "Q or R"

---

### Example: Student Activities

**Given:**
- "Ron is in the Great Hall OR Hermione is in the library"
- "Ron is NOT in the Great Hall OR Harry is sleeping"

**Conclude:**
- "Hermione is in the library OR Harry is sleeping" ‚úì

**Symbolic Notation:**
$$\frac{\text{RonHall} \lor \text{HermioneLib}, \quad \neg\text{RonHall} \lor \text{HarrySleep}}{\text{HermioneLib} \lor \text{HarrySleep}}$$

üì∏ **[IMAGE: Generalized resolution with multiple disjuncts]**

**Step-by-step reasoning:**
1. Either Ron is in hall OR Hermione in library
2. Either Ron NOT in hall OR Harry sleeping
3. The complementary literals $\text{RonHall}$ and $\neg\text{RonHall}$ cancel out
4. We're left with: Hermione in library OR Harry sleeping!

---

## üìù Clauses and Conjunctive Normal Form (CNF)

### Definitions

**Literal**: A propositional symbol or its negation
- Examples: $P$, $\neg Q$, $\text{Raining}$, $\neg\text{Cold}$

**Clause**: A disjunction (Or) of literals
- Examples: 
  - $(P \lor Q \lor R)$
  - $(\neg A \lor B)$
  - $(X \lor \neg Y \lor Z)$

**Conjunctive Normal Form (CNF)**: A conjunction (And) of clauses
- Form: $(C_1) \land (C_2) \land (C_3) \land \ldots$
- Example: $(A \lor B \lor C) \land (D \lor \neg E) \land (F \lor G)$

üì∏ **[IMAGE: Visual representation of literals, clauses, and CNF structure]**

**Why CNF?** Resolution only works on clauses in CNF! All logical statements can be converted to CNF.

---

## üîÑ Converting to Conjunctive Normal Form (CNF)

### The 4-Step Process

**Step 1: Eliminate Biconditionals**
$$\alpha \leftrightarrow \beta \quad \Rightarrow \quad (\alpha \rightarrow \beta) \land (\beta \rightarrow \alpha)$$

**Step 2: Eliminate Implications**
$$\alpha \rightarrow \beta \quad \Rightarrow \quad \neg\alpha \lor \beta$$

**Step 3: Move Negation Inwards** (De Morgan's Laws)
- $\neg(\alpha \land \beta) \Rightarrow \neg\alpha \lor \neg\beta$
- $\neg(\alpha \lor \beta) \Rightarrow \neg\alpha \land \neg\beta$

**Step 4: Distribute Or over And**
$$(\alpha \lor (\beta \land \gamma)) \quad \Rightarrow \quad (\alpha \lor \beta) \land (\alpha \lor \gamma)$$

---

### Detailed Example: Converting $(P \lor Q) \rightarrow R$

**Step 1: Start**
$$(P \lor Q) \rightarrow R$$

**Step 2: Eliminate Implication** (use $\alpha \rightarrow \beta \equiv \neg\alpha \lor \beta$)
$$\neg(P \lor Q) \lor R$$

**Step 3: Apply De Morgan's Law** (move negation inside)
$$(\neg P \land \neg Q) \lor R$$

**Step 4: Apply Distributive Law** (distribute Or over And)
$$(\neg P \lor R) \land (\neg Q \lor R)$$

**Final CNF:** $(\neg P \lor R) \land (\neg Q \lor R)$ ‚úì

üì∏ **[IMAGE: Step-by-step conversion diagram showing transformation]**

---

## üîß Factoring: Removing Duplicates

**Definition**: When resolution produces a clause with duplicate literals, remove the duplicates.

**Example:**
- Clauses: $(P \lor Q \lor S)$ and $(\neg P \lor R \lor S)$
- Resolve on $P$: $(Q \lor S \lor R \lor S)$
- After factoring: $(Q \lor R \lor S)$ ‚úì

**Why?** Because $S \lor S \equiv S$ (idempotent property)

---

## ‚ö†Ô∏è The Empty Clause

**Key Property**: Resolving $P$ and $\neg P$ gives the **empty clause** $()$

**Formula:**
$$\frac{P, \quad \neg P}{()}$$

**The empty clause is ALWAYS FALSE!**

This makes sense: It's impossible for both $P$ and $\neg P$ to be true simultaneously.

**This property enables proof by contradiction!**

üì∏ **[IMAGE: Empty clause derivation showing contradiction]**

---

## üé≠ Proof by Contradiction (Resolution Algorithm)

### The Strategy

**To prove KB ‚ä® Œ±:**
1. Assume the opposite: $\neg\alpha$ (what we want to prove is FALSE)
2. Check if $(KB \land \neg\alpha)$ leads to a contradiction
3. If we derive the empty clause $()$ ‚Üí contradiction found!
4. Therefore, $\alpha$ must be true ‚úì

**Core Idea**: If assuming $\neg\alpha$ creates an impossible situation, then $\alpha$ must be true!

---

### The Algorithm

**To determine if KB ‚ä® Œ±:**

**Step 1:** Convert $(KB \land \neg\alpha)$ to CNF

**Step 2:** Keep applying resolution to produce new clauses

**Step 3:** Check the result:
- If we produce the empty clause $()$ ‚Üí **KB ‚ä® Œ±** ‚úì
- If no more clauses can be inferred ‚Üí **No entailment**

üì∏ **[IMAGE: Flowchart of resolution algorithm]**

---

## üîç Complete Example: Does $(A \lor B) \land (\neg B \lor C) \land (\neg C)$ entail $A$?

### Setup

**KB:** $(A \lor B) \land (\neg B \lor C) \land (\neg C)$

**Query:** $A$

**Goal:** Prove KB ‚ä® $A$ using proof by contradiction

---

### Step-by-Step Proof

**Step 1:** Assume $A$ is false (to prove by contradiction)
$$\text{KB} = (A \lor B) \land (\neg B \lor C) \land (\neg C) \land (\neg A)$$

**Clauses:**
1. $(A \lor B)$
2. $(\neg B \lor C)$
3. $(\neg C)$
4. $(\neg A)$

---

**Step 2:** Resolve clauses (3) and (2) on $C$
$$\frac{(\neg C), \quad (\neg B \lor C)}{(\neg B)}$$

**New clause:** $(\neg B)$

---

**Step 3:** Resolve clauses (1) and new clause on $B$
$$\frac{(A \lor B), \quad (\neg B)}{(A)}$$

**New clause:** $(A)$

---

**Step 4:** Resolve clauses (4) and new clause on $A$
$$\frac{(\neg A), \quad (A)}{()}$$

**Empty clause!** ‚Üí **Contradiction!** ‚úó

---

### Conclusion

Since assuming $\neg A$ led to a contradiction, $A$ must be true!

**Therefore: KB ‚ä® A** ‚úì

üì∏ **[IMAGE: Resolution tree showing the complete derivation]**

---

## üìä Resolution - Summary

### Advantages ‚úÖ

- **Complete**: Can prove any entailment that exists
- **Systematic**: Clear algorithm to follow
- **Efficient**: Often faster than model checking
- **Explanatory**: Shows proof steps

### Process Summary

1. Convert KB and $\neg$(query) to CNF
2. Find complementary literals
3. Apply resolution repeatedly
4. If empty clause ‚Üí query is entailed!
5. If no more inferences ‚Üí not entailed

# First-Order Logic (FOL)

## üéØ Why First-Order Logic?

**Problem with Propositional Logic:**

To express "Every student has an advisor", we'd need separate symbols:
- $\text{HasAdvisor}_{\text{Alice}}$, $\text{HasAdvisor}_{\text{Bob}}$, $\text{HasAdvisor}_{\text{Carol}}$, ...

**Solution: First-Order Logic**

With FOL: $\forall x. \text{Student}(x) \rightarrow \text{HasAdvisor}(x)$

**One sentence for ALL students!** üéâ

## üèóÔ∏è Building Blocks of First-Order Logic

### 1. Constant Symbols (Objects)

**Definition**: Represent specific objects in the world

**Examples:**
- People: $\text{Minerva}$, $\text{Harry}$, $\text{Ron}$
- Places: $\text{Gryffindor}$, $\text{Library}$
- Things: $\text{Wand}$, $\text{Book}$

---

### 2. Predicate Symbols (Properties & Relations)

**Definition**: Functions that take arguments and return True/False

**Unary Predicates** (1 argument - properties):
- $\text{Person}(x)$ ‚Äî "x is a person"
- $\text{House}(x)$ ‚Äî "x is a house"
- $\text{Student}(x)$ ‚Äî "x is a student"

**Binary Predicates** (2 arguments - relationships):
- $\text{BelongsTo}(x, y)$ ‚Äî "x belongs to y"
- $\text{FriendsWith}(x, y)$ ‚Äî "x is friends with y"
- $\text{Teaches}(x, y)$ ‚Äî "x teaches y"

**N-ary Predicates** (multiple arguments):
- $\text{Gave}(x, y, z)$ ‚Äî "x gave y to z"

## üìù First-Order Logic Examples

### Basic Statements

**"Minerva is a person"**
$$\text{Person}(\text{Minerva})$$

**"Gryffindor is a house"**
$$\text{House}(\text{Gryffindor})$$

**"Minerva is NOT a house"**
$$\neg\text{House}(\text{Minerva})$$

**"Minerva belongs to Gryffindor"**
$$\text{BelongsTo}(\text{Minerva}, \text{Gryffindor})$$

---

### Using Logical Connectives

**"If Minerva belongs to Gryffindor, she is not in Slytherin"**
$$\text{BelongsTo}(\text{Minerva}, \text{Gryffindor}) \rightarrow \neg\text{BelongsTo}(\text{Minerva}, \text{Slytherin})$$

## ‚àÄ Universal Quantification

**Symbol:** $\forall$ (upside-down A = "for All")

**Meaning:** "For every x" or "For all x"

**General Form:**
$$\forall x. P(x)$$

**Reads as:** "For all x, P(x) is true"

---

### Example 1: House Exclusivity

**Statement:** "Everyone who belongs to Gryffindor does NOT belong to Hufflepuff"

**FOL:**
$$\forall x. \text{BelongsTo}(x, \text{Gryffindor}) \rightarrow \neg\text{BelongsTo}(x, \text{Hufflepuff})$$

**Breakdown:**
- $\forall x$ ‚Äî For every person x
- $\text{BelongsTo}(x, \text{Gryffindor})$ ‚Äî If x belongs to Gryffindor
- $\rightarrow \neg\text{BelongsTo}(x, \text{Hufflepuff})$ ‚Äî Then x does NOT belong to Hufflepuff

üì∏ **[IMAGE: Universal quantification visualization with domain of objects]**

---

### Example 2: Universal Property

**Statement:** "All students must attend classes"

**FOL:**
$$\forall x. \text{Student}(x) \rightarrow \text{AttendsClasses}(x)$$

**Important Pattern**: Universal quantification often uses implication ($\rightarrow$)
- We don't say "everything in the universe attends classes"
- We say "IF something is a student, THEN it attends classes"

---

## ‚àÉ Existential Quantification

**Symbol:** $\exists$ (backwards E = "Exists")

**Meaning:** "There exists at least one x" or "For some x"

**General Form:**
$$\exists x. P(x)$$

**Reads as:** "There exists an x such that P(x) is true"

---

### Key Difference from Universal

**Universal ($\forall$)**: Makes a claim about EVERY object
**Existential ($\exists$)**: Makes a claim about AT LEAST ONE object

---

### Example 1: Someone Has Property

**Statement:** "Someone is a wizard"

**FOL:**
$$\exists x. \text{Wizard}(x)$$

**Reads as:** There exists (at least one) x such that x is a wizard.

**This is TRUE if:**
- Harry is a wizard, OR
- Hermione is a wizard, OR
- Ron is a wizard, OR
- At least one person is a wizard

---

### Example 2: House Membership

**Statement:** "Minerva belongs to a house" (some house, we don't know which)

**FOL:**
$$\exists x. \text{House}(x) \land \text{BelongsTo}(\text{Minerva}, x)$$

**Reads as:** There exists an x such that (x is a house AND Minerva belongs to x)

**Breakdown:**
- $\exists x$ ‚Äî There exists some x
- $\text{House}(x)$ ‚Äî x is a house
- $\land$ ‚Äî AND
- $\text{BelongsTo}(\text{Minerva}, x)$ ‚Äî Minerva belongs to that x

**Important**: We use AND ($\land$) with existential, not implication!

üì∏ **[IMAGE: Existential quantification showing at least one satisfying object]**

---

### Why AND not Implication?

**WRONG:**
$$\exists x. \text{House}(x) \rightarrow \text{BelongsTo}(\text{Minerva}, x)$$

This would be true even if Minerva doesn't belong to any house! (Because if x is not a house, the implication is vacuously true)

**CORRECT:**
$$\exists x. \text{House}(x) \land \text{BelongsTo}(\text{Minerva}, x)$$

This says: "There is something that IS a house AND Minerva belongs to it"

---

### Pattern Summary

**Universal Quantification**: Use IMPLICATION
$$\forall x. \text{Type}(x) \rightarrow \text{Property}(x)$$
"All things of this type have this property"

**Existential Quantification**: Use AND
$$\exists x. \text{Type}(x) \land \text{Property}(x)$$
"There is at least one thing of this type with this property"

---

## üîó Combining Quantifiers

### Example 1: Every Person Belongs to a House

**Statement:** "Every person belongs to (at least) one house"

**FOL:**
$$\forall x. \text{Person}(x) \rightarrow \left(\exists y. \text{House}(y) \land \text{BelongsTo}(x, y)\right)$$

**Breakdown:**
- $\forall x$ ‚Äî For every x
- $\text{Person}(x) \rightarrow$ ‚Äî If x is a person, then...
- $\exists y$ ‚Äî There exists a y such that...
- $\text{House}(y) \land \text{BelongsTo}(x, y)$ ‚Äî y is a house AND x belongs to y

**Natural language:** "For any person you pick, you can find a house they belong to"

üì∏ **[IMAGE: Combined quantifiers showing nested relationships]**

---

### Example 2: Friendship

**Statement:** "Everyone has at least one friend"

**FOL:**
$$\forall x. \text{Person}(x) \rightarrow \left(\exists y. \text{Person}(y) \land \text{FriendsWith}(x, y)\right)$$

## üéØ Comparison: Propositional vs First-Order Logic

### House Assignment Problem

**Propositional Logic:**
- Need 16 separate symbols:
  - $\text{MinervaGryffindor}$, $\text{MinervaHufflepuff}$, $\text{MinvervaRavenclaw}$, $\text{MinervaSlytherin}$
  - $\text{PomonaGryffindor}$, ... (12 more)
- Need ~100 rules to encode constraints
- Very verbose and repetitive

**First-Order Logic:**
- Constants: $\text{Minerva}$, $\text{Pomona}$, $\text{Gryffindor}$, $\text{Hufflepuff}$
- Predicate: $\text{BelongsTo}(x, y)$
- One rule: $\forall x, y, z. \text{Person}(x) \land \text{BelongsTo}(x, y) \land \text{BelongsTo}(x, z) \rightarrow (y = z)$
  - "Each person belongs to at most one house"

**Much more concise!** ‚ú®

## üîÑ Quantifier Equivalences

### De Morgan's Laws for Quantifiers

**Negating Universal:**
$$\neg\forall x. P(x) \equiv \exists x. \neg P(x)$$

**Example:** "Not everyone passed" $\equiv$ "Someone failed"

**Negating Existential:**
$$\neg\exists x. P(x) \equiv \forall x. \neg P(x)$$

**Example:** "No one passed" $\equiv$ "Everyone failed"
