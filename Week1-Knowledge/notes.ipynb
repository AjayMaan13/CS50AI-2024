## 1. Introduction: Reasoning from Knowledge

Humans use **knowledge** to reason and draw new conclusions.
AI systems can do the same ‚Äî these are called **Knowledge-Based Agents**.

> üß© Example:
>
> * If it didn‚Äôt rain, Harry visited Hagrid today.
> * Harry visited Hagrid or Dumbledore today, but not both.
> * Harry visited Dumbledore today.
>
> From these, we can reason:
> ‚Üí Harry did **not** visit Hagrid.
> ‚Üí Therefore, **it rained today**.

AI represents and manipulates such knowledge using **logic**.

---

## 2. üßæ Sentences and Propositional Logic

### ‚û§ Sentence

An **assertion** about the world, expressed in a knowledge representation language.

### ‚û§ Propositional Logic

A way of representing facts about the world using **propositions**, which can be **True (T)** or **False (F)**.

---

## 3. üî† Propositional Symbols

Usually capital letters (`P`, `Q`, `R`) represent simple propositions.

Example:

* P: ‚ÄúIt is raining.‚Äù
* Q: ‚ÄúI am indoors.‚Äù

---

## 4. üîó Logical Connectives

| Symbol | Name                   | Meaning              | True When                  |
| :----- | :--------------------- | :------------------- | :------------------------- |
| ¬¨      | **Not**                | Negates truth value  | Opposite of P              |
| ‚àß      | **And**                | Both must be true    | P and Q both true          |
| ‚à®      | **Or**                 | At least one true    | P or Q or both true        |
| ‚äï      | **Exclusive Or (XOR)** | Exactly one true     | Only P or only Q           |
| ‚Üí      | **Implication**        | ‚ÄúIf P, then Q‚Äù       | False only when P=T, Q=F   |
| ‚Üî      | **Biconditional**      | ‚ÄúP if and only if Q‚Äù | True when P and Q are same |

---

## 5. ‚ùó Truth Tables

### (a) **NOT (¬¨)**

|  P  |  ¬¨P |
| :-: | :-: |
|  T  |  F  |
|  F  |  T  |

---

### (b) **AND (‚àß)**

|  P  |  Q  | P ‚àß Q |
| :-: | :-: | :---: |
|  F  |  F  |   F   |
|  F  |  T  |   F   |
|  T  |  F  |   F   |
|  T  |  T  |   T   |

---

### (c) **OR (‚à®)**

|  P  |  Q  | P ‚à® Q |
| :-: | :-: | :---: |
|  F  |  F  |   F   |
|  F  |  T  |   T   |
|  T  |  F  |   T   |
|  T  |  T  |   T   |

üëâ *Inclusive* OR allows both to be true.
üëâ *Exclusive* OR (XOR) is true only if **exactly one** is true.

---

## 6. ‚û°Ô∏è **Implication (‚Üí)**

### P ‚Üí Q means:

> ‚ÄúIf P, then Q.‚Äù

üìä **Truth Table for Implication**

|  P  |  Q  | P ‚Üí Q | Explanation                                                                                         |
| :-: | :-: | :---: | :-------------------------------------------------------------------------------------------------- |
|  T  |  T  | **T** | If P is true and Q is true, the rule ‚Äúif P then Q‚Äù is satisfied.                                    |
|  T  |  F  | **F** | If P is true but Q is false, the rule is broken ‚Äî this is the only case where implication is false. |
|  F  |  T  | **T** | If P is false, we don‚Äôt care what Q is ‚Äî the rule isn‚Äôt violated.                                   |
|  F  |  F  | **T** | Same here ‚Äî if P is false, ‚Äúif P then Q‚Äù doesn‚Äôt get tested, so it‚Äôs considered true.               |

### üí° Example

If P = ‚ÄúIt‚Äôs raining‚Äù and Q = ‚ÄúI‚Äôm indoors‚Äù,
then **P ‚Üí Q** means ‚ÄúIf it‚Äôs raining, then I‚Äôm indoors.‚Äù

---

## 7. ‚Üî **Biconditional (If and Only If)**

### P ‚Üî Q

or
‚Äúif and only if‚Äù
OR
**P if and only if Q**

üß† **What It Means**
A biconditional says:

> ‚ÄúP and Q have the same truth value.‚Äù

üìä **Truth Table for Biconditional**

|  P  |  Q  | P ‚Üî Q | Explanation              |
| :-: | :-: | :---: | :----------------------- |
|  T  |  T  | **T** | Both are true ‚Üí match ‚úÖ  |
|  T  |  F  | **F** | Don‚Äôt match ‚ùå            |
|  F  |  T  | **F** | Don‚Äôt match ‚ùå            |
|  F  |  F  | **T** | Both are false ‚Üí match ‚úÖ |

### üí° Example

If P = ‚ÄúIt‚Äôs raining‚Äù and Q = ‚ÄúI‚Äôm indoors,‚Äù
then **P ‚Üî Q** means:

> ‚ÄúIf it‚Äôs raining, then I‚Äôm indoors, and if I‚Äôm indoors, then it‚Äôs raining.‚Äù

---

## 8. üß© Model

A **model** is a specific assignment of truth values to all propositions.

Example:

* P: ‚ÄúIt‚Äôs raining.‚Äù
* Q: ‚ÄúIt‚Äôs Tuesday.‚Äù

One possible model:
‚Üí {P = True, Q = False}

With 2 propositions, there are 2¬≤ = 4 possible models.

---

## 9. üß† Knowledge Base (KB)

A **knowledge base** is a collection of sentences (facts, rules) known to the AI.

Example:

```
1. If it didn‚Äôt rain, Harry visited Hagrid.
2. Harry visited Hagrid or Dumbledore, but not both.
3. Harry visited Dumbledore.
```

From these, KB can infer:
‚Üí It rained today.

---

## 10. ‚ä® **Entailment**

**Entailment (‚ä®)** means:

> If Œ± ‚ä® Œ≤, then in *every model* where Œ± is true, Œ≤ is also true.

### Example:

* Œ±: ‚ÄúIt‚Äôs Tuesday in January.‚Äù
* Œ≤: ‚ÄúIt‚Äôs January.‚Äù
  Then Œ± ‚ä® Œ≤ (every world where Œ± is true, Œ≤ is also true).

üß© Difference:

* **Implication (‚Üí)** is a **logical connective** inside sentences.
* **Entailment (‚ä®)** is a **relationship** between two sentences.

# Propositional Logic: Inference & Knowledge Engineering

## üìñ Part 1: Inference

### What is Inference?
**Inference** is the process of deriving new sentences (conclusions) from old ones (premises).

**Example**: In Harry Potter logic:
- From sentences 1, 2, and 3 ‚Üí we can infer sentences 4 and 5

---

## üîç Model Checking Algorithm

### Definition
To determine if **KB ‚ä® Œ±** (Does the knowledge base entail Œ±?):

**Algorithm Steps:**
1. **Enumerate all possible models** (all truth value combinations)
2. **Check each model**: If KB is true in that model, is Œ± also true?
3. **Conclusion**: If Œ± is true in EVERY model where KB is true ‚Üí KB ‚ä® Œ±

---

### üìä Example Problem

**Given:**
- **P**: It is a Tuesday
- **Q**: It is raining
- **R**: Harry will go for a run

**Knowledge Base (KB):**
1. `(P ‚àß ¬¨Q) ‚Üí R` ‚Äî "If it's Tuesday AND not raining, Harry goes for a run"
2. `P` ‚Äî "It IS Tuesday"
3. `¬¨Q` ‚Äî "It is NOT raining"

**Query**: `R` ‚Äî Does Harry go for a run?

### Step-by-Step Evaluation

#### Step 1: Apply `P` is true
Since P must be true, KB is **false** in all models where P = false

#### Step 2: Apply `¬¨Q` (Q is false)
Since Q must be false, KB is **false** in all models where Q = true

#### Step 3: Apply `(P ‚àß ¬¨Q) ‚Üí R`
When P = true AND Q = false, R must be true
- Model where R = false ‚Üí KB is **false**
- Model where R = true ‚Üí KB is **true** ‚úì

## üíª Code Representation

### Example: Harry Potter Logic Puzzle

```python
from logic import *

# Create symbols for each proposition
rain = Symbol("rain")           # It is raining
hagrid = Symbol("hagrid")       # Harry visited Hagrid
dumbledore = Symbol("dumbledore")  # Harry visited Dumbledore

# Build the Knowledge Base
knowledge = And(
    Implication(Not(rain), hagrid),      # ¬¨rain ‚Üí hagrid
    Or(hagrid, dumbledore),              # hagrid ‚à® dumbledore
    Not(And(hagrid, dumbledore)),        # ¬¨(hagrid ‚àß dumbledore)
    dumbledore                           # dumbledore is TRUE
)
```

**What the KB says:**
1. If it's not raining, Harry visited Hagrid
2. Harry visited Hagrid OR Dumbledore (at least one)
3. Harry did NOT visit both Hagrid and Dumbledore
4. Harry DID visit Dumbledore (fact)

## üîÑ The check_all Function

### Required Inputs
1. **Knowledge Base** ‚Äî Used to draw inferences
2. **Query** ‚Äî The proposition we want to check
3. **Symbols** ‚Äî List of all atomic propositions (variables)
4. **Model** ‚Äî Assignment of truth values to symbols

---

### üñºÔ∏è **[IMAGE 3: Model Checking Flowchart]**
*Visual representation of the recursive model checking process*

### Algorithm Code

```python
def check_all(knowledge, query, symbols, model):
    
    # Base case: all symbols have been assigned
    if not symbols:
        # If KB is true in this model
        if knowledge.evaluate(model):
            return query.evaluate(model)  # Check if query is true
        return True  # KB is false ‚Üí vacuous truth
    
    # Recursive case: still have symbols to assign
    else:
        remaining = symbols.copy()
        p = remaining.pop()  # Pick one symbol
        
        # Create two models
        model_true = model.copy()
        model_true[p] = True
        
        model_false = model.copy()
        model_false[p] = False
        
        # Recurse on both and combine with AND
        return (check_all(knowledge, query, remaining, model_true) and
                check_all(knowledge, query, remaining, model_false))
```

---

### üéØ Key Insight: Why "Vacuous Truth"?

**We only care about models where KB is true!**

**Example**: 
- **KB**: P (Harry plays seeker), Q (Oliver plays keeper), (P ‚àß Q) ‚Üí R (Gryffindor wins)
- **Model**: ¬¨P (Harry plays beater instead)

In this model, we don't care about R because our KB says P is true. This model is **irrelevant** to our knowledge.

**If KB is false in a model ‚Üí return True (ignore this model)**

## üéÆ Example 1: Clue/Cluedo Game

### Game Setup
**Murder mystery with:**
- **People**: Mustard, Plum, Scarlet
- **Tools**: Knife, Revolver, Wrench
- **Locations**: Ballroom, Kitchen, Library

**Goal**: Determine WHO did it, with WHAT tool, in WHICH location

---

### Building the Knowledge Base

#### Step 1: Game Rules
One person, one tool, one location must be true:

```python
knowledge = And(
    # Exactly one person is guilty
    Or(mustard, plum, scarlet),
    
    # Exactly one tool was used
    Or(knife, revolver, wrench),
    
    # Exactly one location
    Or(ballroom, kitchen, library)
)
```

---

#### Step 2: Initial Cards (What We See)
We see: Mustard, Kitchen, Revolver
Therefore, these are NOT the answer:

```python
knowledge = And(
    # ... previous rules ...
    Not(mustard),
    Not(kitchen),
    Not(revolver)
)
```

---

#### Step 3: False Guess
Someone guesses: "Scarlet, Library, Wrench" ‚Äî **WRONG**
This means: At least ONE of these is false

```python
knowledge = And(
    # ... previous knowledge ...
    Or(Not(scarlet), Not(library), Not(wrench))
)
```

**Translation**: ¬¨Scarlet ‚à® ¬¨Library ‚à® ¬¨Wrench
(At least one must be false)

---

#### Step 4: More Clues
Someone shows us Plum card:

```python
knowledge = And(
    # ... previous knowledge ...
    Not(plum)
)
```

**Now we can deduce**: Since Mustard and Plum are out ‚Üí **Scarlet is guilty!**

---

#### Step 5: Final Clue
We learn: NOT the Ballroom

```python
knowledge = And(
    # ... previous knowledge ...
    Not(ballroom)
)
```

---

### üß© Final Deduction

**What we know:**
1. Must be Ballroom, Kitchen, or Library
2. NOT Ballroom, NOT Kitchen
3. Therefore: **Library ‚úì**

**From the false guess:**
- Scarlet ‚úì, Library ‚úì, Wrench ?
- Since at least one must be false, and we know Scarlet and Library are TRUE
- Therefore: **Wrench is FALSE**, so it must be **Knife ‚úì**

**Solution: Scarlet, Knife, Library** üéâ

---

### Complete Clue Code (`clue.py`)

```python
import termcolor
from logic import *

# Define symbols for characters, rooms, and weapons
mustard = Symbol("ColMustard")
plum = Symbol("ProfPlum")
scarlet = Symbol("MsScarlet")
characters = [mustard, plum, scarlet]

ballroom = Symbol("ballroom")
kitchen = Symbol("kitchen")
library = Symbol("library")
rooms = [ballroom, kitchen, library]

knife = Symbol("knife")
revolver = Symbol("revolver")
wrench = Symbol("wrench")
weapons = [knife, revolver, wrench]

symbols = characters + rooms + weapons

def check_knowledge(knowledge):
    """Check what we can deduce about each symbol"""
    for symbol in symbols:
        if model_check(knowledge, symbol):
            # We can prove this symbol is TRUE
            termcolor.cprint(f"{symbol}: YES", "green")
        elif not model_check(knowledge, Not(symbol)):
            # We can't prove it's false, so it's uncertain
            print(f"{symbol}: MAYBE")
        # If neither condition is true, it's proven FALSE (print nothing)

# Build the knowledge base
knowledge = And(
    # Game rules: one person, one room, one weapon
    Or(mustard, plum, scarlet),
    Or(ballroom, kitchen, library),
    Or(knife, revolver, wrench)
)

# Initial cards we saw (these are NOT the answer)
knowledge.add(And(
    Not(mustard), Not(kitchen), Not(revolver)
))

# Someone's wrong guess (at least one is false)
knowledge.add(Or(
    Not(scarlet), Not(library), Not(wrench)
))

# Additional cards revealed
knowledge.add(Not(plum))
knowledge.add(Not(ballroom))

# Check what we can deduce
check_knowledge(knowledge)
```

**Output:**
```
MsScarlet: YES     (proven guilty!)
library: YES       (proven location!)
knife: YES         (proven weapon!)
```

**Key Feature**: The `check_knowledge()` function has three outcomes:
- **YES** (green): We can prove it's true ‚Üí `model_check(KB, symbol)` returns True
- **MAYBE**: We can't prove it's false ‚Üí `model_check(KB, Not(symbol))` returns False
- **Nothing printed**: We can prove it's false ‚Üí Both checks return False

## üéì Example 2: House Assignment Puzzle (`puzzle.py`)

### Problem
Four people (Gilderoy, Pomona, Minerva, Horace) assigned to four houses (Gryffindor, Hufflepuff, Ravenclaw, Slytherin). Each person in exactly one house.

---

### Complete Implementation

```python
from logic import *

people = ["Gilderoy", "Pomona", "Minerva", "Horace"]
houses = ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]

symbols = []
knowledge = And()

# Create all 16 propositions (4 people √ó 4 houses)
for person in people:
    for house in houses:
        symbols.append(Symbol(f"{person}{house}"))

# CONSTRAINT 1: Each person belongs to a house
for person in people:
    knowledge.add(Or(
        Symbol(f"{person}Gryffindor"),
        Symbol(f"{person}Hufflepuff"),
        Symbol(f"{person}Ravenclaw"),
        Symbol(f"{person}Slytherin")
    ))

# CONSTRAINT 2: Only one house per person
for person in people:
    for h1 in houses:
        for h2 in houses:
            if h1 != h2:
                # If person is in h1, they're NOT in h2
                knowledge.add(
                    Implication(Symbol(f"{person}{h1}"), 
                              Not(Symbol(f"{person}{h2}")))
                )

# CONSTRAINT 3: Only one person per house
for house in houses:
    for p1 in people:
        for p2 in people:
            if p1 != p2:
                # If p1 is in this house, p2 is NOT
                knowledge.add(
                    Implication(Symbol(f"{p1}{house}"), 
                              Not(Symbol(f"{p2}{house}")))
                )

# CLUE 1: Gilderoy is in Gryffindor or Ravenclaw
knowledge.add(
    Or(Symbol("GilderoyGryffindor"), Symbol("GilderoyRavenclaw"))
)

# CLUE 2: Pomona is NOT in Slytherin
knowledge.add(
    Not(Symbol("PomonaSlytherin"))
)

# CLUE 3: Minerva IS in Gryffindor
knowledge.add(
    Symbol("MinervaGryffindor")
)

# Find all true assignments
for symbol in symbols:
    if model_check(knowledge, symbol):
        print(symbol)
```

**Output Example:**
```
MinervaGryffindor    (given as fact)
GilderoyRavenclaw    (deduced from clues)
PomonaHufflepuff     (deduced by elimination)
HoraceSlytherin      (deduced by elimination)
```

### Challenge: Representing in Propositional Logic

**Need 16 propositions:**
- MinervaGryffindor, MinervaHufflepuff, MinervaRavenclaw, MinervaSlytherin
- PomonaGryffindor, PomonaHufflepuff, ...
- (4 people √ó 4 houses = 16 propositions)

---

### Constraints to Encode

#### 1. Each person belongs to A house:
```python
# For each person, at least one house must be true
Or(MinervaGryffindor, MinervaHufflepuff, 
   MinervaRavenclaw, MinervaSlytherin)
```
Repeat for each person (4 times)

#### 2. Each person belongs to ONLY ONE house:
```python
# For each pair of houses, if person in h1, NOT in h2
for h1 in houses:
    for h2 in houses:
        if h1 != h2:
            Implication(Symbol(f"{person}{h1}"), 
                       Not(Symbol(f"{person}{h2}")))
```
For each person: 12 implications (4 √ó 3)
Total: 48 implications for all 4 people

#### 3. Each house has ONLY ONE person:
```python
# For each pair of people, if p1 in house, p2 NOT in house
for p1 in people:
    for p2 in people:
        if p1 != p2:
            Implication(Symbol(f"{p1}{house}"), 
                       Not(Symbol(f"{p2}{house}")))
```
For each house: 12 implications (4 √ó 3)
Total: 48 more implications for all 4 houses

**Total Constraints**: 4 (at least one house) + 48 (one house per person) + 48 (one person per house) = **100 logical statements**

**Problem**: Very cumbersome! This is why **First-Order Logic** is useful (covered later).

## üé® Example 3: Mastermind Game (`mastermind.py`)

### Game Rules
- Player 1: Arranges colors in secret order
- Player 2: Guesses the order
- Player 1: Returns number of correct positions

---

### Complete Implementation

```python
from logic import *

colors = ["red", "blue", "green", "yellow"]
symbols = []

# Create all propositions: color + position
for i in range(4):
    for color in colors:
        symbols.append(Symbol(f"{color}{i}"))

knowledge = And()

# CONSTRAINT 1: Each color has a position
for color in colors:
    knowledge.add(Or(
        Symbol(f"{color}0"),
        Symbol(f"{color}1"),
        Symbol(f"{color}2"),
        Symbol(f"{color}3")
    ))

# CONSTRAINT 2: Only one position per color
for color in colors:
    for i in range(4):
        for j in range(4):
            if i != j:
                knowledge.add(Implication(
                    Symbol(f"{color}{i}"), 
                    Not(Symbol(f"{color}{j}"))
                ))

# CONSTRAINT 3: Only one color per position
for i in range(4):
    for c1 in colors:
        for c2 in colors:
            if c1 != c2:
                knowledge.add(Implication(
                    Symbol(f"{c1}{i}"), 
                    Not(Symbol(f"{c2}{i}"))
                ))

# CLUE 1: First guess - exactly 2 positions correct
# red0, blue1, green2, yellow3 ‚Üí 2 correct
# This means exactly 2 of these are true
knowledge.add(Or(
    And(Symbol("red0"), Symbol("blue1"), 
        Not(Symbol("green2")), Not(Symbol("yellow3"))),
    And(Symbol("red0"), Symbol("green2"), 
        Not(Symbol("blue1")), Not(Symbol("yellow3"))),
    And(Symbol("red0"), Symbol("yellow3"), 
        Not(Symbol("blue1")), Not(Symbol("green2"))),
    And(Symbol("blue1"), Symbol("green2"), 
        Not(Symbol("red0")), Not(Symbol("yellow3"))),
    And(Symbol("blue1"), Symbol("yellow3"), 
        Not(Symbol("red0")), Not(Symbol("green2"))),
    And(Symbol("green2"), Symbol("yellow3"), 
        Not(Symbol("red0")), Not(Symbol("blue1")))
))

# CLUE 2: Second guess - 0 correct
# blue0, red1, green2, yellow3 ‚Üí all wrong
knowledge.add(And(
    Not(Symbol("blue0")),
    Not(Symbol("red1")),
    Not(Symbol("green2")),
    Not(Symbol("yellow3"))
))

# Find the solution
for symbol in symbols:
    if model_check(knowledge, symbol):
        print(symbol)
```
